```
# Q3(a)
import numpy as np
import matplotlib.pyplot as plt
n = 15
t = np.arange(0,1,0.001)
h = np.exp(-5*(t-3)**2)+0.5*np.exp(-100*(t-5)**2)+0.5*np.exp(-100*(t-0.75)**2)
line_up,=plt.plot(t,h,label='True Function')
x_i = np.random.uniform(size=n)
mu=0
sigma=1
w = np.random.normal(mu,sigma,n)
t = x_i
y_i = np.exp(-5*(t-3)**2)+0.5*np.exp(-100*(t-5)**2)+0.5*np.exp(-100*(t-0.75)**2)+w
line_down,=plt.plot(x_i,y_i,'ro',label='Noisy Observation')
plt.title('True Function and Noisy Observation')
plt.legend(handles=[line_up, line_down])
plt.xlabel('t or X_i')
plt.ylabel('h(t) or Y_i')
```

```
# Q3(b)
w = np.array([1,-2,3])
print(w)
temp = np.arange(0,1,0.001)
t_polyd2 = np.array([np.ones(len(temp)),temp,temp**2])
h_polyd2 = w.dot(t_polyd2)
print(h_polyd2.shape)
plt.plot(t_polyd2[2,:],h_polyd2)
plt.title('Degree 2 Polynomial h(t) with w=[1,-2,3]')
plt.xlabel('t')
plt.ylabel('h(t)')
```
